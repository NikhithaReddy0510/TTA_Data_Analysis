```python
# Asking user to pick a number and store against a variable called number.
user_number = int(input("Please enter some number between 1 and 100: " ))

# Creating an IF statement which uses logical operators which prints out a joke based on the number the user picked
if user_number >= 1 and user_number <= 20:
    print("your number is " + str(user_number) + " So, first Joke: My wife told me to stop impersonating a flamingo. I had to put my foot down.")
elif user_number >= 21 and user_number <= 50:
    print("your number is " + str(user_number) + " So, second Joke: I was wondering why the frisbee kept getting bigger and bigger, but then it hit me.")
elif user_number >= 51 and user_number <= 100:
    print("your number is " + str(user_number) + " So, third Joke: Don’t you hate it when someone answers their own questions? I do.")
else :
    print("Number is out of range")
```

# Task 4 
A motorbike costs £2000 and loses 10% of its value every year. Using a loop, print the value of the bike every following year until it falls below £1000.

```python
# Storing the value of the bike against a variable initial_bike_value.
initial_bike_value = 2000

purchase_year = 0

# Storing lose percentage of bike in a variable depreciation.
depreciation = 0.1

# To calculate the current value of bike.
current_bike_value = initial_bike_value 

# Creating a while loop which subtracts 10% of the bikes value untill the value reaches £1000.
while current_bike_value > 1000 :
    print("In year " +str(purchase_year)+" the value of bike is: " +str(current_bike_value))
    current_bike_value = current_bike_value - (current_bike_value * depreciation)
    purchase_year += 1
```

# Task 5 
Write a program which will ask for two numbers from a user. Then offer a menu to the user giving them a choice of operator:e.g. – Enter “a” if you want to add“b” if you want to subtract. Include +, -, /, , * square (to the power of). Once the user has selected which operator they wish to use, perform the calculation.

```python
# Asking the user to pick a number and store in a variable user_first_number.
user_first_number = int(input("Please select your first number: "))

# Asking the user to pick a second number and store in a variable user_second_number.
user_second_number = int(input("Please select your second number: "))

# Asking the user to pick a operator and store in a variable user_operator. 
user_operator = input("Choose your operator (+, -, /, *, **, %, //) : ")

# Creating an IF statement to carry out the math operation using the users choice of numbers and choice of operator.
if user_operator == '+' :
    print( user_first_number + user_second_number)
elif user_operator == '-' :
    print(user_first_number - user_second_number)
elif user_operator == '/' :
    print(user_first_number / user_second_number)
elif user_operator == '*' :
    print(user_first_number * user_second_number)
elif user_operator == '**' :
    print(user_first_number ** user_second_number)
elif user_operator == '%' :
    print(user_first_number % user_second_number)
elif user_operator == '//' :
    print(user_first_number // user_second_number)
else:
    print("An appropriate operator was not selected!")
```
